```python
import pandas as pd
```

```python
dfs = []
for i in range(0, 4):
    df = pd.read_excel("Data.xlsx", sheet_name=i)
    df.columns = df.columns.str.strip()
    # Loại bỏ khoảng trắng ở đầu và cuối của dữ liệu
    df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
    df = df.map(lambda x: x.title() if isinstance(x, str) else x)
    dfs.append(df)
    
```

# Kiểm tra các giá trị null(hoặc nan) trong từng bảng

```python
for df in dfs:
    print(df.isna().sum())
```

Bảng "Nhân viên" xuất hiện các giá trị nan, chúng ta sẽ xử lí các giá trị này

```python
# lấy ra bảng đầu tiên ("Nhân viên")
df = dfs[0]
```

```python
# tổng quan về bảng dữ liệu
df.info()
df
```

# Xử lí dữ liệu

```python
# Các giá trị nan trong cột "Kinh nghiệm trong ngành", chúng ta sẽ để giá trị = -1, sau đó ép kiểu dữ liệu cho cột này thành int
df["Kinh nghiệm trong ngành"] = df["Kinh nghiệm trong ngành"].fillna(-1).astype(int)
df
```

## Xem xét một chút 2 cột "Kinh nghiệm làm việc" và "Kinh nghiệm trong ngành": ta thấy rằng số năm kinh nghiệm trong ngành luôn phải lớn hơn hoặc bằng số năm kinh nghiệm làm việc!

```python
# tìm ra các bản ghi có dữ liệu không phù hợp với điều kiện đã đưa ra
mask = (df["Kinh nghiệm trong ngành"] < df["Kinh nghiệm làm việc tại Viện"])
```

```python
# Các bản ghi không phù hợp
df[mask]
```

```python
# Sửa lỗi (nếu có): Gán giá trị hợp lý cho "Kinh nghiệm làm việc trong ngành"
df.loc[mask, "Kinh nghiệm trong ngành"] = df.loc[mask,"Kinh nghiệm làm việc tại Viện"]
```

```python
df
```

```python
# kiểm tra lại, ta thấy không còn bản ghi nào không phù hợp với điều kiện đã đưa ra
df[df["Kinh nghiệm trong ngành"] < df["Kinh nghiệm làm việc tại Viện"]]
```

## Tương tự như cột "Kinh nghiệm làm việc", ta cũng sẽ thay thế các giá trị nan trong cột "Tuổi" = -1 và ép kiểu dữ liệu trong cột này thành "int"

```python
df['Tuổi']=df["Tuổi"].fillna(-1).astype(int)
```

## Đặt Vấn đề với cột "Tuổi":
### Đầu tiên, ta sẽ tạo một cột mới có tên là "Tuổi bắt đầu làm việc" để tính toán độ tuổi mà một nhân viên bắt đầu làm việc(hành nghề) bằng cách lấy số Tuổi - Kinh nghiệm trong ngành.
### Dựa trên cột "Nơi cư trú", ta sẽ giả định các nhân viên này đều làm việc tại Nga. Theo đó, theo luật lao động ở Nga, các chức vụ trong cột "Loại" sẽ có số tuổi được quy định để bắt đầu công việc:
####   'Nhân Viên Phục Vụ': 18 -> đủ 18 tuổi mới được nhận công việc này
####   'Nhân Viên Y Tế': 18 -> đủ 18 tuổi hơn
####   'Bác Sĩ': 22 -> đủ 22 tuổi hoặc hơn
####   'Sự Quản Lý': 25 -> đủ 25 hoặc hơn

## So sánh các giá trị trong cột "Tuổi bắt đầu làm việc" với các giá trị được quy định trên, ta sẽ thấy có những bản ghi không phù hợp, kéo theo các giá trị trong cột "Tuổi" của các bản ghi này cũng sẽ có vấn đề (cột "Kinh nghiệm trong ngành" đã được chỉnh sửa trước)

```python
min_age_by_position = {
    'Nhân Viên Phục Vụ': 18,
    'Nhân Viên Y Tế': 18,
    'Bác Sĩ': 22,
    'Sự Quản Lý': 25
}

# Áp dụng tuổi tối thiểu vào một cột mới
df['Tuổi tối thiểu hợp lý'] = df['Loại'].map(min_age_by_position)

# Kiểm tra các dòng có vi phạm (tuổi hiện tại - số kinh nghiệm trong ngành < tuổi tối thiểu hợp lý)
mask_age_violation = (df['Tuổi'] -df["Kinh nghiệm trong ngành"]) < df['Tuổi tối thiểu hợp lý']
mask_age_violation = mask_age_violation.reindex(df.index)


print("Các dòng có dữ liệu không hợp lệ:")
print(df[mask_age_violation][df["Loại"] =="Bác Sĩ"])
print(df[mask_age_violation][df["Loại"] =="Nhân Viên Y Tế"])
print(df[mask_age_violation][df["Loại"] =="Nhân Viên Phục Vụ"])
print(df[mask_age_violation][df["Loại"] =="Sự Quản Lý"])
```

```python
# cập nhật lại dữ liệu trong cột "Tuổi"
df.loc[mask_age_violation, "Tuổi"] = df.loc[mask_age_violation, "Tuổi tối thiểu hợp lý"] + df.loc[mask_age_violation, "Kinh nghiệm trong ngành"]
```

```python
# kiểm tra lại các giá trị trong cột "Tuổi", "Tuổi tối thiểu hợp lý" và "Kinh nghiệm trong ngành"
df
```

```python
x = df["Nơi cư trú"].mode()[0]
x
df["Nơi cư trú"].fillna(x, inplace = True)
df["Nơi cư trú"].isna().sum()
```
